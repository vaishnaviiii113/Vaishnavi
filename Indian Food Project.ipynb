# Now I'm going to use pie chart for veg & non veg food distirbution

```python
df[['course','prep_time']].sort_values(by='prep_time',ascending=False).reset_index().head()
```

From this we get to know that main course takes highest time for preparation

```python

x= df[['flavor_profile','prep_time']].groupby(['flavor_profile']).agg(['max','mean','min'])
x.plot(kind='line')
plt.legend()
plt.show()
```

Now we are going to add cook time and prep time to find the total time needed to serve a dish

```python
df['total_time'] = df['prep_time']+df['cook_time']

df.loc[df.groupby(['course'])['total_time'].idxmax()][['total_time','course']]
```

```python
df[['total_time','course']].groupby(['course']).agg(['max','min']).plot(kind='bar', color=['brown','green'])
plt.title("Time taken to be served", fontsize=20)
plt.ylabel("Time",fontsize=15)
plt.show()
```
